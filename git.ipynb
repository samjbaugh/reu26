# Git Helper

This notebook lets you do basic git operations from within leaving Jupyter.

**High-level Overview:**
There are three stages to pushing your changes to a git repository: 
1. Add new files (git add)
2. Commit to your **local repository** (git commit)
3. Push to the **remote repository** (git push)
Before doing these steps, it is usually a good idea to first do a git pull, and to also do a "git status" to check which files have been changed.

To obtain files from the repository (i.e., ones that I add) you can do **git pull**. 

**Warning**: If the remote repository and the local repository have different changes to the same file, this will lead to a merge conflict, which can be a bit of a pain. To avoid this, let's just try not to edit the same files!


---
## 1. Check status

This shows which files you've changed, added, or deleted since your last commit.
- **Modified** = you edited an existing file
- **Untracked** = a new file git doesn't know about yet


```python
!git status
```

---
## 2. See what you changed

This shows a line-by-line diff of your modifications.
Lines starting with `+` are additions, `-` are deletions.

Note: this only shows changes to files git already tracks (not new/untracked files).

```python
!git diff
```

## 6. View recent history

See the last 10 commits: who made them, when, and the message.

```python
!git log --oneline --format="%h  %an  %ar  %s" -10
```

---
## 7. View a specific commit

Copy a commit hash (the short code like `f8780e0`) from the log above and paste it below to see what files were changed.

```python
# Paste a commit hash from the log above
COMMIT_HASH = "f8780e0"

!git show --stat {COMMIT_HASH}
print("\n" + "="*60 + "\n")
!git show {COMMIT_HASH}
```

## 8. See who last changed each line of a file

We probably won't need this too much, but this is a useful way to see "when" (in terms of the git commit history) a change occured, and who made the change:

```python
# Change the filename to whichever file you're curious about
FILENAME = "README.md"

!git blame --date=short {FILENAME}
```
