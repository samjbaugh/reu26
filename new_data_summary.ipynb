```python
import pandas as pd
import numpy as np

def summarize(df, name):
    """Print a clean summary of a dataframe."""
    print(f"{'='*60}")
    print(f"  {name}")
    print(f"{'='*60}")
    print(f"  Shape: {df.shape[0]} rows x {df.shape[1]} columns")
    print(f"  Columns: {list(df.columns)}")
    if 'year' in df.columns:
        print(f"  Year range: {int(df['year'].min())} - {int(df['year'].max())}")
    print()
    # Per-column summary
    for col in df.columns:
        if col == 'year':
            continue
        if df[col].dtype in ['float64', 'int64', 'float32']:
            vals = df[col].dropna()
            print(f"  {col}:")
            print(f"    non-null: {len(vals)} / {len(df)}")
            print(f"    range: [{vals.min():.3f}, {vals.max():.3f}]")
            print(f"    mean: {vals.mean():.3f}  std: {vals.std():.3f}")
        else:
            print(f"  {col}: {df[col].nunique()} unique values ({df[col].dtype})")
    print()
```

## 2. Precipitable Water (PWAT)

**Source:** NCEP/NCAR Reanalysis 1 (Kalnay et al., 1996)  
**Resolution:** Monthly, 2.5-degree grid, 1948-present  
**Units:** kg/m^2 (equivalent to mm of liquid water if the entire column condensed)  

**Regional averages** (for exploration):

| Column | Region | Lat | Lon |
|--------|--------|-----|-----|
| `pwat_eastern_us` | Entire eastern US | 25-48N | 100-65W |
| `pwat_southeast_us` | Southeast US | 25-37N | 90-75W |
| `pwat_northeast_us` | Northeast US | 37-48N | 85-65W |
| `pwat_gulf_coast` | Gulf coast | 25-32N | 98-80W |

**Station-level values:** In the full predictor table, each station also gets its own
`pwat_station` value interpolated from the grid (nearest grid point). This is important
because precipitable water varies a lot spatially -- coastal and southern stations have
much more atmospheric moisture than inland/northern ones.


```python
pwat = pd.read_csv("data/pwat_regional.csv")
summarize(pwat, "Precipitable Water (NCEP/NCAR Reanalysis)")
pwat.head()
```

## 3. Distance to Water Bodies


| Column | Description |
|--------|-------------|
| `dist_coast_km` | Distance to nearest large water body (Atlantic, Gulf, OR Great Lakes) |
| `dist_atlantic_km` | Distance to nearest Atlantic/Gulf coast point |
| `dist_greatlakes_km` | Distance to nearest Great Lakes shoreline point |

Uses Haversine distance to the nearest point on a detailed coastline approximation (~40 points for Atlantic/Gulf, ~30 for Great Lakes).


```python
# Demonstrate the distance calculation on the 12 taster stations
import sys
sys.path.insert(0, "..")

# Station locations from the taster dataset
taster_stations = {
    "Boston Logan, MA":    (42.36, -71.01),
    "Central Park, NY":    (40.78, -73.97),
    "Miami, FL":           (25.79, -80.29),
    "Chicago OHare, IL":   (41.99, -87.91),
    "Atlanta, GA":         (33.77, -84.39),
    "Pittsburgh, PA":      (40.49, -80.23),
    "Burlington, VT":      (44.47, -73.15),
    "Nashville, TN":       (36.12, -86.69),
    "Portland, ME":        (43.65, -70.31),
    "Richmond, VA":        (37.51, -77.32),
    "Raleigh-Durham, NC":  (35.89, -78.78),
    "New Orleans, LA":     (29.98, -90.25),
}

# Reuse the haversine function
import numpy as np

def haversine_km(lon1, lat1, lon2, lat2):
    R = 6371.0
    lon1, lat1, lon2, lat2 = map(np.radians, [lon1, lat1, lon2, lat2])
    dlat = lat2 - lat1
    dlon = lon2 - lon1
    a = np.sin(dlat/2)**2 + np.cos(lat1)*np.cos(lat2)*np.sin(dlon/2)**2
    return R * 2 * np.arcsin(np.sqrt(a))

ATLANTIC_COAST = np.array([
    (-80.1, 25.8), (-80.2, 26.7), (-80.1, 28.0), (-80.5, 29.0),
    (-81.3, 30.3), (-81.1, 31.2), (-80.9, 32.0), (-79.9, 32.8),
    (-78.9, 33.8), (-77.9, 34.2), (-76.5, 34.7), (-75.5, 35.2),
    (-75.8, 36.0), (-75.5, 37.0), (-76.0, 37.5), (-75.5, 38.3),
    (-74.0, 38.9), (-74.0, 39.5), (-74.0, 40.5), (-73.0, 40.7),
    (-72.0, 41.2), (-71.4, 41.5), (-70.6, 41.5), (-70.0, 41.7),
    (-70.2, 42.0), (-70.8, 42.3), (-70.0, 43.5), (-67.0, 44.8),
    (-81.8, 24.5), (-82.6, 27.8), (-83.5, 29.0), (-84.3, 30.0),
    (-85.5, 29.7), (-86.5, 30.4), (-87.5, 30.3), (-88.5, 30.2),
    (-89.5, 29.2), (-90.5, 29.5), (-91.5, 29.5), (-93.8, 29.7),
    (-95.0, 29.0), (-97.2, 26.1),
])

GREAT_LAKES = np.array([
    (-76.2, 43.5), (-77.3, 43.3), (-78.0, 43.3), (-79.0, 43.3),
    (-79.5, 42.8), (-80.0, 42.5), (-81.5, 42.0), (-82.5, 41.7),
    (-83.0, 41.7), (-83.5, 41.7),
    (-82.5, 43.0), (-83.0, 42.3), (-83.5, 44.0), (-84.0, 44.8),
    (-84.5, 45.5), (-84.7, 45.8),
    (-87.0, 45.0), (-86.5, 44.0), (-86.8, 43.0), (-87.0, 42.0),
    (-87.8, 42.0), (-86.5, 41.7), (-87.5, 44.5),
    (-84.5, 46.5), (-86.0, 46.5), (-87.0, 47.0), (-88.5, 47.0),
    (-89.5, 47.0), (-90.5, 47.0), (-92.0, 46.8),
])

rows = []
for name, (lat, lon) in taster_stations.items():
    d_atl = np.min(haversine_km(lon, lat, ATLANTIC_COAST[:, 0], ATLANTIC_COAST[:, 1]))
    d_lake = np.min(haversine_km(lon, lat, GREAT_LAKES[:, 0], GREAT_LAKES[:, 1]))
    rows.append({"station": name, "dist_atlantic_km": round(d_atl, 1),
                 "dist_greatlakes_km": round(d_lake, 1),
                 "dist_coast_km": round(min(d_atl, d_lake), 1)})

dist_df = pd.DataFrame(rows).sort_values("dist_coast_km")
print(dist_df.to_string(index=False))
```
